```python
from flask import Flask, request, render_template_string
import math

app = Flask(__name__)

@app.route('/sqrt',methods=['GET'])
def compte_sqrt():
    try:
        num1 = request.args.get('num1',type=float)
        num2 = request.args.get('num2',type=float)

        if num1 is None or num2 is None:
            return render_template_string('<h1>Error: Both num1 & num2 are required</h2>'),400
        
        if num2==0:
            return render_template_string('<h1>Error: Divison by zero</h1>'),400
        
        divison_result = num1/num2

        if divison_result<0:
            return render_template_string('<h1>Can not compute sqrt of negative number</h1>'),400
        
        sqrt_result = math.sqrt(divison_result)

        return render_template_string(
            '<h1> Square root of {{divison_result}} is {{sqrt_result}}</h1>',
            divison_result = divison_result,
            sqrt_result = sqrt_result
        )
    except ValueError:
        return render_template_string(
            '<h1> Error: Invalid input. Please provide numeric values'
        )
if __name__ == '__main__':
    app.run(debug=False)
```
